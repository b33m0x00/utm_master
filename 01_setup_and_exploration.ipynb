## Download HarmBench Dataset

```python
dataset = load_dataset('walledai/HarmBench', 'standard')

dataset['train'][:10]
```

## Save HarmBench Dataset

```python
data = []

for i, example in enumerate(dataset['train']):
    data.append({
        'id': f'harmbench_{i}',
        'prompt': example['prompt'],
        'type': 'jailbreak',
        'category': example['category'],
        'source': 'HarmBench'
    })

with open('datasets/raw/harmbench.json', 'w') as f:
    json.dump(data, f, indent=2)
```
